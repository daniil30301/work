```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
import pickle
import json
```

```python
Ovechkin_18 = json.load(open('ovechkin18.json'))
Ovechkin_19 = json.load(open('ovechkin19.json'))
Ovechkin_17 = json.load(open('ovechkin17.json'))
Ovechkin_16 = json.load(open('ovechkin16.json'))
Ovechkin_15 = json.load(open('ovechkin15.json'))
```

```python
def goals(player_goals_season):
    
    player_x_goal_normalized = []
    player_y_goal_normalized = []

    for i, s in enumerate(player_goals_season['Goal']['x']):
        if player_goals_season['Goal']['x'][i] < 0:
            player_x_goal_normalized.append(-player_goals_season['Goal']['x'][i])
            player_y_goal_normalized.append(-player_goals_season['Goal']['y'][i])
        else:
            player_x_goal_normalized.append(player_goals_season['Goal']['x'][i])
            player_y_goal_normalized.append(player_goals_season['Goal']['y'][i])
    goal = {}
    goal['Goal'] = {}
    goal['Goal']['x'] = []
    goal['Goal']['y'] = []
    goal['Goal']['x'].append(player_x_goal_normalized)
    goal['Goal']['y'].append(player_y_goal_normalized)
    
    return goal
```

```python
ovi_19 = goals(Ovechkin_19)
ovi_18 = goals(Ovechkin_18)
ovi_17 = goals(Ovechkin_17)
ovi_16 = goals(Ovechkin_16)
ovi_15 = goals(Ovechkin_15)
```

```python
fig = plt.figure(figsize = (15,11))
ax1 = fig.add_subplot(3,3,1)
ax2 = fig.add_subplot(3,3,2)
ax3 = fig.add_subplot(3,3,3)
ax4 = fig.add_subplot(3,3,4)
ax5 = fig.add_subplot(3,3,5)
# ax1.imshow(I)

ax1.scatter(ovi_19['Goal']['x'], ovi_19['Goal']['y'], color = 'r')
ax2.scatter(ovi_18['Goal']['x'], ovi_18['Goal']['y'], color = 'g')
ax3.scatter(ovi_17['Goal']['x'], ovi_17['Goal']['y'], color = 'y')
ax4.scatter(ovi_16['Goal']['x'], ovi_16['Goal']['y'], color = 'b')
ax5.scatter(ovi_15['Goal']['x'], ovi_15['Goal']['y'])

ax1.set_title('goals')
ax2.set_title('goals')
ax3.set_title('goals')
ax4.set_title('goals')
ax5.set_title('goals')

ax1.set_yticks([-50, -40, -30, -20, -10, 0 , 10, 20, 30, 40, 50])
ax2.set_yticks([-50, -40, -30, -20, -10, 0 , 10, 20, 30, 40, 50])
ax3.set_yticks([-50, -40, -30, -20, -10, 0 , 10, 20, 30, 40, 50])
ax4.set_yticks([-50, -40, -30, -20, -10, 0 , 10, 20, 30, 40, 50])
ax5.set_yticks([-50, -40, -30, -20, -10, 0 , 10, 20, 30, 40, 50])
ax1.set_xticks([10, 20, 30, 40, 50, 60, 70, 80, 90])
ax2.set_xticks([10, 20, 30, 40, 50, 60, 70, 80, 90])
ax3.set_xticks([10, 20, 30, 40, 50, 60, 70, 80, 90])
ax4.set_xticks([10, 20, 30, 40, 50, 60, 70, 80, 90])
ax5.set_xticks([10, 20, 30, 40, 50, 60, 70, 80, 90])
```
